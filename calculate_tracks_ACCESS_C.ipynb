```python
def gen_tracks(tracks_obj, filenames, b_path=None):

    grids = (
        pyart.io.read_grid(fn, include_fields = ['reflectivity'])
        for fn in filenames)
    
      
    tracks_obj.get_tracks(grids, b_path=b_path)
        
    return tracks_obj
```
