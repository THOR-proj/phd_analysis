```python
# Core 
import datetime
import os
import glob

# Analysis 
import xarray as xr
import numpy as np
import pyproj as pp
import scipy as sp
import metpy as mp

# Plotting 
import cartopy.crs as ccrs
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
from matplotlib import rcParams
import matplotlib.ticker as mticker
from matplotlib.animation import FuncAnimation
import matplotlib.colors as colors


# Debugging 
import pdb
%load_ext line_profiler
```

```python
CPOL_20 = xr.open_mfdataset('/g/data/rr5/CPOL_radar/CPOL_level_1b/GRIDDED/GRID_70km_1000m/2016/20161220/*.nc')
```

```python
CPOL_20 = CPOL_20[['reflectivity','velocity']].load()
```

```python
plt.close('all')

CPOL_i = CPOL_20.sel(x=slice(-70000,15000), z=slice(0,12000))
CPOL_i = CPOL_i.sel(y=0, method='nearest')

theta = np.arctan2(CPOL_i.z, CPOL_i.x)
v = - np.sin(theta) * CPOL_i.velocity
u = - np.cos(theta) * CPOL_i.velocity

for t in range(117,127):
    plt.figure()
       
    plt.pcolormesh(
        CPOL_i.x, CPOL_i.z, CPOL_i.reflectivity.isel(time=t), 
        vmin=-30, vmax=80, cmap='gist_ncar'
    )
    plt.colorbar()
    
    plt.quiver(
        CPOL_i.x[::2], CPOL_i.z[::2], u.isel(time=t)[::2, ::2], v.isel(time=t)[::2, ::2],
        units='xy', angles='xy', 
    )
```
